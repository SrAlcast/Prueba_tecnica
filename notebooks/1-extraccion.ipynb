```python
list_type="short","movie"
list_year=list(range(1990,2025))
list_genres=["Drama", "Comedy", "Action", "Fantasy", "Horror", "Mystery", "Romance", "Thriller"]
```

```python
def buscar_peliculas(Year,titleType,genre,page):
    url = "https://moviesdatabase.p.rapidapi.com/titles"
    headers = {
    "X-RapidAPI-Key": API_KEY, 
    "X-RapidAPI-Host": "moviesdatabase.p.rapidapi.com"}
    
    params = {    
    "year": Year,           # Año 
    "titleType": titleType, # Tipo 
    "genre":genre,          # Filtrar titulo
    "page":page,
    "limit":50}            # Pagina resultado
    
    response = requests.get(url, headers=headers, params=params)
    data = response.json()  # Convertir respuesta a JSON
    pagej = data.get("page")  # Extraer los resultados
    movies = data.get("results", [])  # Extraer los resultados
    lista_titles=[]
    for movie in movies:
        title = (
            movie["id"],
            movie["titleText"]["text"],
            movie["titleType"]["text"],
            movie["releaseYear"]["year"]
              )
        lista_titles.append(title)
    return lista_titles
```

```python
all_titles1=[]
for type in tqdm(list_type):
    for genre in tqdm(list_genres):
        for year in tqdm(list_year):
                lista_titulos=buscar_peliculas(year,type,genre,1)
                all_titles1.extend(lista_titulos)
```
