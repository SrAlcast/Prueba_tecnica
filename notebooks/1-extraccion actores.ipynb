```python
def buscar_peliculas(page):
    url = "https://moviesdatabase.p.rapidapi.com/actors"
    headers = {
    "X-RapidAPI-Key": API_KEY, 
    "X-RapidAPI-Host": "moviesdatabase.p.rapidapi.com"}

    params = {    
    "page":page,
    "limit":50}            # Pagina resultados
    
    response = requests.get(url, headers=headers, params=params)
    data = response.json()  # Convertir respuesta a JSON
    actors = data.get("results", [])  # Extraer los resultados
    lista_actors=[]
    for actor in actors:
        title = (
            actor["nconst"],
            actor["primaryName"],
            actor["primaryProfession"],
            actor["knownForTitles"]
              )
        lista_actors.append(title)
    return lista_actors
```

```python
all_actores=[]
list_pages=list(range(1,50))
for page in tqdm(list_pages):
    lista_actores=buscar_peliculas(page)
    all_actores.extend(lista_actores)
```
